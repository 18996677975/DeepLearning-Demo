# 数据准备

```python
x = data.iloc[:, 1:-1]
y = data.iloc[:, -1]
```

```python
x.shape
```

# 搭建神经网络

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(200, input_shape=(3,), activation='relu'),
                             tf.keras.layers.Dense(100, activation='relu'),
                             tf.keras.layers.Dense(50, activation='relu'),
                             tf.keras.layers.Dense(25, activation='relu'),
                            tf.keras.layers.Dense(1)])

model.summary()
```

# 设置优化模型、损失函数

```python
model.compile(optimizer='adam', loss='mse')
```

# 模型训练

```python
history = model.fit(x, y, epochs=2000)
```

# 准备测试集

```python
test = data.iloc[:10, 1:-1]

test
```

# 以训练的模型对测试集进行预测

```python
model.predict(test)
```
